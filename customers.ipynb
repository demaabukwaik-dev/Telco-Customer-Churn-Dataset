phase 5

```python
# feature 1 :
# extract new feature is_month_to_month from Contract column 
# it indicates the customers contract type (month or not) which is expected to affect churn analysis
# is_month_to_month 1 if Contract == "Month-to-month"  else 0
df["is_month_to_month"] = df["Contract"].apply(lambda x : 1 if x == "Month-to-month" else 0  )
```

```python
# feature 2 :
# i can extract a more specific feature than is_month_to_month indicating the contract length group (month / year / two-year)
# extract new feature contract_length_group from Contract column
# it indicates the customers contract type (month or year or two-year) which is expected to affect churn analysis
def contract_length_group (contract) :
    if contract == "Month-to-month" :
        return "month"
    elif contract == "One year" :
        return "year"
    else:
        return "two-year"
df["contract_length_group"] = df["Contract"].apply(contract_length_group)
```

```python
# feature 3 :
# extract avg_spend_per_month feature from TotalCharges and tenure columns
# avg_spend_per_month = TotalCharges / (tenure + 1)   --> add 1 to tenure to avoid division by zero for new customers (tenure = 0)
# TotalCharges accumulates over time, while this represents the actual monthly value regardless of the duration the customer spent in the company
# this allows fair comparison between new and long-term customers, 
# it helps analyze churn by comparing past average with current price
df["avg_spend_per_month"] = df["TotalCharges"] / ( df["tenure"] + 1 )
```

```python
# feature 4 :
# extract new feature high_monthly_charge from MonthlyCharges column
# high_monthly_charge 1 if MonthlyCharges > median else 0 
def monthly_charge_seg (charge) :
        if charge  > df["MonthlyCharges"].median():
            return "high"
        else:
            return "low"
df["high_monthly_charge"] = df["MonthlyCharges"].apply(monthly_charge_seg)
```

```python
# feature 5 :
# extract new feature high_total_charge from TotalCharges column
# high_total_charge 1 if TotalCharges > median else 0
median_charge = df["TotalCharges"].median()
df["high_total_charge"] = df["TotalCharges"].apply( lambda x : "high" if x > median_charge else "low" )
```

```python
# feature 6 :
# extract new feature new_customer from tenure column
# new_customer 1 if tenure <= 3 else 0
df["new_customer"] = df["tenure"].apply(lambda x : x<=3)
```

```python
# feature 7 :
# extract new feature long_term_customer from tenure column
# long_term_customer 1 if tenure >= 24 else 0
df["long_term_customer"] = df["tenure"].apply(lambda x : x>=24 )
```

```python
# check  --> must be no intersection 
df[df["new_customer"] & df ["long_term_customer"] ]  
```

```python
# feature 8 :
# extract new feature uses_electronic_check from PaymentMethod column
# uses_electronic_check 1 if PaymentMethod == "Electronic check" else 0
df["uses_electronic_check"] = df["PaymentMethod"].apply(lambda x : 1 if x == "Electronic check" else 0 )
```

```python
# feature 10 :
# extract new feature has_internet_service from InternetService column
# has_internet_service 1 if InternetService != "No" else 0
df["has_internet_service"] = df["InternetService"]!= "No"
```

```python
# feature 11 :
# extract new feature Diff between MonthlyCharges and avg_spend_per_month to analyze the change in the customers spending behavior
# Diff = MonthlyCharges - avg_spend_per_month
df["Diff"] = df["MonthlyCharges"] - df["avg_spend_per_month"]
```

```python
# feature 12 :
# extract new feature price_changed from Diff column to analyze the effect of sudden price changes on churn
# if the diff is positive and high (>10) --> this means a sudden price increase happens --> likely to churn
# if the diff is negative                --> this means a sudden price decrease happens --> likely to stay
df["price_changed"] = df["Diff"].apply(lambda x : 1 if x > 10 else 0 )
```

```python
# feature 13 : 
# extract new feature is_automatic from PaymentMethod column
# indicate whether the customer uses an automatic payment method (bank transfer or credit card) or not (electronic check or mailed check)
# is_automatic 1 if PaymentMethod is (Bank transfer (automatic) or Credit card (automatic)) else 0
df["is_automatic"] = df["PaymentMethod"].apply(lambda x : 1 if x in ["Bank transfer (automatic)","Credit card (automatic)"] else 0 )    
```

```python
# feature 14 :
# extract new feature payment_category from PaymentMethod column
# it indicates the payment method type (automatic) or not (electronic check or mailed check) which is expected to affect churn analysis
# why grouping payment methods into 2 categories (automatic vs non-automatic) is not accurate ? 
# because customers who pay with electronic check or mailed check have different churn rates as shown below in df.groupby("PaymentMethod")["Churn"].mean()
# for some analysis its not correct to ignore this difference
def payment_category(method):
    if 'automatic' in method:
        return 'Automatic'
    else:
        return method 

df['payment_category'] = df['PaymentMethod'].apply(payment_category)
df.groupby("payment_category")["Churn"].mean()
```

Feature Scaling
Scaling Approach:
For tenure, avg_spend_per_month, and MonthlyCharges, as there are no obvious outliers and the distributions seems balanced, StandardScaler will be applied to normalize the values (mean = 0, variance = 1)

For Diff, due to the presence of some extreme values, RobustScaler will be used. This scaler relies on the median & IQR to reduce the impact of outliers while preserving the information they represent.
